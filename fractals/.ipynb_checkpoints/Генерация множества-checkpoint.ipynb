#### Генерация ряда

```python
import pandas as pd
import numpy as np
import plotly
import plotly.graph_objs as go
import decimal
```

```python
file = "set_kantor_probabl.txt"
data = pd.read_csv(file, sep=';', header=None)
x = np.round(data[1].values, decimals = 3)
y = [1 for i in range(len(x))]
```

```python
file = "set_serpinskii_probabl.txt"
data = pd.read_csv(file, sep=';', header=None)
x = np.round(data[1].values, decimals = 3)
y = np.round(data[2].values, decimals = 3)
```

```python
len(x)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y,mode='markers'))
fig.update_layout(margin=dict(l=0, r=0, t=0, b=0))
fig.show()
```

```python
fig.write_html("set_kantor_probabl.html")
```
